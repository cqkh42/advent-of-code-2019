```python
maze = ["#########","#b.A.@.a#","#########"]
maze = [
        "#################",
        "#i.G..c...e..H.p#",
        "########.########",
        "#j.A..b...f..D.o#",
        "########@########",
        "#k.E..a...g..B.n#",
        "########.########",
        "#l.F..d...h..C.m#",
        "#################"
    ]

#a, f, b, j, g, n, h, d, l, o, e, p, c, i, k, m
```

```python
maze_dict = {}
positions = {}
for y, line in enumerate(maze):
    for x, val in enumerate(line):
        maze_dict[(x, y)] = val
        if val.islower() or val == "@":
            positions[val] = (x, y)
```

```python
NUM_KEYS = len(positions) - 1
```

```python
# which keys can I reach from "@"
```

```python
from functools import lru_cache
```

```python
start = positions["@"]
keys = "0" * NUM_KEYS
```

```python
@lru_cache(maxsize=None)
def keys_available(position, keys):
    queue = [[position, keys, 0]]
    visited = set((position, keys))
    can_find = {}
    
    while len(queue):
        location, keys, length = queue.pop(0)
#         print(f"at {location}")
        if (location, keys) in visited:
            continue
        else:
            visited.add((location, keys))
    
        next_bits = (
                (location[0]+1, location[1]),
                (location[0]-1, location[1]),
                (location[0], location[1]+1),
                (location[0], location[1]-1)
            )
        next_bits = {i for i in next_bits if maze_dict[i] != "#"}
        next_bits = next_bits.difference(visited)
        for those_coords in next_bits:
            val_there = maze_dict[those_coords]
            if val_there.islower():
#                 print(f"found {val_there}")
                new_keys = keys[:ord(val_there)-97]+"1"+keys[ord(val_there)-97+1:]
                if (val_there, new_keys) not in can_find:
                    can_find[(val_there, new_keys)] = length+1
            elif val_there.isupper():
#                 print(f"hit {val_there}")
                if keys[ord(val_there)-65] == "0":
                    continue
            
            queue.append([those_coords, keys, length+1])
    return can_find
```

```python

ALL = {}
queue = [[start, keys]]
i = 0
while len(queue):
    i += 1
    if i > 10000:
        break
    a, b = queue.pop(0)
    available = keys_available(a, b)
    x = {((maze_dict[a], b), k): v for k, v in available.items()}
    
    ALL.update(x)
    queue.extend((positions[a], b) for a, b in available)
```
